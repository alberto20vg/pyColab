E7. Mostrar las máximas diferencias entre temperatura máxima y mínima por provincia. Para ellos habrá que crear un campo calculado que podemos llamar “diff_temp_min_max”. Mostrar solo las columnas fecha, estación, provincia, temp_min, temp_max y diff_temp_min_max.


```
df['diff_temp_min_max']=df['temp_max']-df['temp_min']
df.iloc[df.reset_index().groupby('provincia')['diff_temp_min_max'].idxmax()].sort_values(by='diff_temp_min_max',ascending=False)[['fecha','estacion','provincia','diff_temp_min_max','temp_min','temp_max']]
```

E8. Mostrar las mayores rachas de viento en un día por provincia ordenadas descendentemente. En la tabla de salida se deben mostrar las siguientes columnas: fecha, estación, provincia y viento_racha.
